```python
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap
```

```python
best_rf_fraud = joblib.load("../models/random_forest_fraud.pkl")
best_rf_credit = joblib.load("../models/random_forest_credit.pkl")
```

```python
fraud_df = pd.read_csv("../data/processed/fraud_processed.csv")
credit_df = pd.read_csv("../data/processed/creditcard_processed.csv")
```

```python
Xf = fraud_df.drop(columns=["class"])
yf = fraud_df["class"]
```

```python
Xc = credit_df.drop(columns=["Class"])
yc = credit_df["Class"]
```

```python
Xf = Xf.select_dtypes(include=[np.number])
Xc = Xc.select_dtypes(include=[np.number])
```

```python
print(Xf.shape, Xc.shape)
```

```python
Xf_sample = Xf.sample(100, random_state=42)
Xc_sample = Xc.sample(100, random_state=42)
```

```python
explainer_fraud = shap.TreeExplainer(best_rf_fraud)
explainer_credit = shap.TreeExplainer(best_rf_credit)
```

```python
shap_values_fraud = explainer_fraud.shap_values(Xf_sample)
```

```python
shap_values_credit = explainer_credit.shap_values(Xc_sample)
```

```python
shap_fraud = shap_values_fraud[1]
shap_credit = shap_values_credit[1]
```

```python
shap.summary_plot(shap_fraud, Xf_sample)
```

```python
shap.summary_plot(shap_credit, Xc_sample)
```

Due to the computational complexity of SHAP on large datasets, a stratified sample of the data was used for model interpretation. This approach preserves the distribution of the original dataset while significantly reducing computation time, without affecting the reliability of global feature importance analysis.
