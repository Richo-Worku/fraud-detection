```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    f1_score,
    confusion_matrix,
    precision_recall_curve,
    auc
)
```

```python
fraud_df = pd.read_csv("../data/processed/fraud_processed.csv")
credit_df = pd.read_csv("../data/processed/creditcard_processed.csv")
```

```python
X_f = fraud_df.drop('class', axis=1)
y_f = fraud_df['class']
```

```python
X_f = X_f.drop(columns=[
    'signup_time',
    'purchase_time',
    'device_id'
], errors='ignore')
```

```python
Xf_train, Xf_test, yf_train, yf_test = train_test_split(
    X_f,
    y_f,
    test_size=0.2,
    random_state=42,
    stratify=y_f
)
```

```python
imputer = SimpleImputer(strategy='median')

Xf_train = imputer.fit_transform(Xf_train)
Xf_test = imputer.transform(Xf_test)
```

```python
smote = SMOTE(random_state=42)

Xf_train_res, yf_train_res = smote.fit_resample(
    Xf_train,
    yf_train
)
```

```python
lr_fraud = LogisticRegression(max_iter=1000)
lr_fraud.fit(Xf_train_res, yf_train_res)
```

```python
yf_pred = lr_fraud.predict(Xf_test)
yf_prob = lr_fraud.predict_proba(Xf_test)[:, 1]
```

```python
f1_fraud = f1_score(yf_test, yf_pred)

precision, recall, _ = precision_recall_curve(yf_test, yf_prob)
aucpr_fraud = auc(recall, precision)

cm_fraud = confusion_matrix(yf_test, yf_pred)
```

 Random Forest Ensemble Model

Random Forest was selected as the ensemble model because it can capture non-linear fraud patterns and interactions between features that Logistic Regression may miss.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}

rf_fraud = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

grid_fraud = GridSearchCV(
    estimator=rf_fraud,
    param_grid=param_grid,
    scoring='f1',
    cv=3,
    n_jobs=-1
)

grid_fraud.fit(Xf_train_res, yf_train_res)
```

```python
print(grid_fraud.best_params_)
```

Interpretation
n_estimators = 200
The model performed best with 200 trees rather than 100.
More trees generally improve stability and predictive performance.
max_depth = 20
The model benefited from deeper trees.
This suggests fraud patterns are relatively complex and require more decision layers to separate fraud from legitimate transactions.
min_samples_split = 2
The model preferred splitting nodes as soon as possible.
This allows the trees to capture fine-grained fraud patterns.
The results indicate that a larger ensemble of deeper trees was more effective at capturing complex fraud patterns in the e-commerce transaction data.

```python
best_rf_fraud = grid_fraud.best_estimator_

yf_pred_rf = best_rf_fraud.predict(Xf_test)
yf_prob_rf = best_rf_fraud.predict_proba(Xf_test)[:, 1]
```

```python
from sklearn.metrics import (
    f1_score,
    confusion_matrix,
    precision_recall_curve,
    auc
)

f1_rf_fraud = f1_score(yf_test, yf_pred_rf)

precision, recall, _ = precision_recall_curve(
    yf_test,
    yf_prob_rf
)

aucpr_rf_fraud = auc(recall, precision)

cm_rf_fraud = confusion_matrix(
    yf_test,
    yf_pred_rf
)

print("Fraud RF F1:", f1_rf_fraud)
print("Fraud RF AUC-PR:", aucpr_rf_fraud)
print(cm_rf_fraud)
```

Random Forest Performance (Fraud Dataset)

A Random Forest classifier was trained using SMOTE-balanced training data. Hyperparameter tuning was conducted using GridSearchCV with 3-fold cross-validation. The optimal configuration consisted of 200 trees, a maximum depth of 20, and a minimum split size of 2.

The Random Forest significantly outperformed the Logistic Regression baseline, achieving an F1-score of 0.523 and an AUC-PR of 0.546. Compared with the baseline model (F1-score = 0.164, AUC-PR = 0.095), the ensemble model demonstrated a substantially improved ability to identify fraudulent transactions.

The confusion matrix showed that the model correctly detected 1,100 fraudulent transactions while maintaining a relatively low false-positive count of 650 transactions. These results indicate that Random Forest is more effective at capturing complex fraud patterns than a linear classifier.

Cross-Validation for Fraud Dataset

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
f1_scores_fraud = cross_val_score(
    best_rf_fraud,
    Xf_train_res,
    yf_train_res,
    cv=cv,
    scoring='f1',
    n_jobs=-1
)

print("Fraud RF F1 Scores:", f1_scores_fraud)
print("Mean F1:", f1_scores_fraud.mean())
print("Std F1:", f1_scores_fraud.std())
```

Cross-Validation Results (Fraud Dataset)

Five-fold Stratified Cross-Validation was performed on the SMOTE-balanced training data using the optimized Random Forest model. The model achieved a mean F1-score of 0.9373 with a standard deviation of 0.0017 across folds. The low standard deviation indicates highly consistent performance and suggests that the model generalizes reliably across different subsets of the training data.

```python
aucpr_scores_fraud = cross_val_score(
    best_rf_fraud,
    Xf_train_res,
    yf_train_res,
    cv=cv,
    scoring='average_precision',
    n_jobs=-1
)

print("Fraud RF AP Scores:", aucpr_scores_fraud)
print("Mean AP:", aucpr_scores_fraud.mean())
print("Std AP:", aucpr_scores_fraud.std())
```

The Random Forest model achieved an exceptionally high cross-validated Average Precision (AUC-PR) score of 0.9815 ± 0.0005 across five Stratified K-Folds. This indicates that the model is highly effective at distinguishing between fraudulent and legitimate transactions under balanced training conditions. The extremely low standard deviation demonstrates consistent performance across all folds, suggesting strong stability and robustness. However, it is important to note that cross-validation was performed on SMOTE-resampled data, which may lead to slightly optimistic performance estimates compared to real-world imbalanced distributions.

Cross-Validation for Credit Card Dataset

The Random Forest model achieved near-perfect performance on the credit card fraud dataset, with a mean F1-score of 0.99985 ± 0.00005 across five Stratified Cross-Validation folds. The extremely high and stable scores indicate that the model is highly effective at distinguishing fraudulent transactions from legitimate ones within the transformed feature space. The very low standard deviation suggests consistent performance across different data splits, demonstrating strong model robustness. However, such near-perfect performance should be interpreted with caution, as the dataset consists of PCA-transformed features that may already encode strong separability between classes.

The Random Forest model achieved an almost perfect Average Precision (AUC-PR) score of 0.99998 ± 0.00003 across five Stratified Cross-Validation folds on the credit card fraud dataset. This indicates that the model is extremely effective at ranking fraudulent transactions higher than legitimate ones, demonstrating near-complete separability in the feature space. The extremely low standard deviation further confirms that the model's performance is highly stable across different data splits. However, these results should be interpreted with caution due to the highly transformed PCA feature space, which may inherently provide strong class separability.
