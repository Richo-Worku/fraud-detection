```python
fraud_df['ip_address_int'] = fraud_df['ip_address'].astype(int)
```

```python
ip_df.dtypes
```

```python
ip_df['lower_bound_ip_address'] = ip_df['lower_bound_ip_address'].astype(int)
ip_df['upper_bound_ip_address'] = ip_df['upper_bound_ip_address'].astype(int)
```

```python
ip_df = ip_df.sort_values('lower_bound_ip_address')
```

```python
fraud_df_sorted = fraud_df.sort_values('ip_address_int')
```

```python
merged_df = pd.merge_asof(
    fraud_df_sorted,
    ip_df,
    left_on='ip_address_int',
    right_on='lower_bound_ip_address',
    direction='backward'
)
```

```python
merged_df = merged_df[
    merged_df['ip_address_int'] <= merged_df['upper_bound_ip_address']
]
```

```python
merged_df = merged_df.sort_index()
```

```python
fraud_df = merged_df
```

```python
fraud_df.drop(
    ['lower_bound_ip_address', 'upper_bound_ip_address'],
    axis=1,
    inplace=True
)
```

```python
fraud_df['country'].value_counts().head(10)
```

```python
fraud_df[['ip_address', 'ip_address_int', 'country']].head()
```

A range-based join was performed between transaction IP addresses and IP address geolocation ranges. Since IP addresses do not have direct keys, a merge-as-of strategy was used to efficiently map each transaction to its corresponding country based on IP range boundaries.


```python
country_fraud = fraud_df.groupby('country')['class'].agg(['count', 'sum'])
```

```python
country_fraud['fraud_rate'] = (
    country_fraud['sum'] / country_fraud['count']
) * 100
```

```python
country_fraud_sorted = country_fraud.sort_values(
    by='fraud_rate',
    ascending=False
)

country_fraud_sorted.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_countries = country_fraud_sorted.head(10)

plt.figure(figsize=(12,6))
sns.barplot(
    x=top_countries.index,
    y=top_countries['fraud_rate']
)

plt.xticks(rotation=45)
plt.title("Top Countries by Fraud Rate")
plt.ylabel("Fraud Rate (%)")
plt.show()
```

The countries with the highest observed fraud rates include:

Turkmenistan
Namibia
Sri Lanka
Luxembourg
Virgin Islands (U.S.)
Ecuador
Tunisia
Peru
Bolivia
Kuwait
Turkmenistan exhibited the highest fraud rate among the countries analyzed.

The results suggest that transaction location may provide valuable information for fraud detection. Certain countries exhibit significantly higher proportions of fraudulent transactions than others, indicating potential geographic risk patterns.
